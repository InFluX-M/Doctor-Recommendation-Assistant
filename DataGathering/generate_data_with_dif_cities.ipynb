```python
import pandas as pd
import numpy as np
```

```python
cities = pd.read_csv('data/top-cities.csv', header=None)
```

```python
cities
```

```python
data_gpt = pd.read_csv("./data/chatgpt_data_generated.csv")
data_form = pd.read_csv("./data/form_data.csv", index_col=0)
```

```python
data_gpt['label'] = data_gpt['label'].apply(lambda x: x.replace(" ", ","))
```

```python
data = pd.concat([data_form, data_gpt], ignore_index=True)
```

```python
data = data[['sentence', 'label']]
```

```python
data
```

```python
datas = []
```

```python
for index, city in cities.iterrows():
    data = data.sample(frac=1).reset_index(drop=True)
    city_sen = city[0].split()
    labels = ['B-cty']
    for u in range(len(city_sen) - 1):
        labels.append('I-cty')
    random_rows = data.sample(n=100, random_state=index)
    for _, row in random_rows.iterrows():
        words, tags = row['sentence'].split(), row['label'].split(',')
        f, l = 0, 0
        if tags.count('B-cty') != 1 or tags.count('B-nhd') != 0:
            continue
        for i in range(len(words)):
            if tags[i] == 'B-cty':
                f = i
                l = i
            if tags[i] == 'I-cty':
                l = i
        del words[f:l+1]
        del tags[f:l+1]
        words[f:f] = city_sen
        tags[f:f] = labels
        datas.append((' '.join(words), ','.join(tags)))
        break
```

```python
data_new = pd.DataFrame(datas, columns=['sentence', 'label'])
```
